```python
stream = ollama.chat(
    model='llama3.2:1b',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)

newstream = ollama.chat(
    model='llama3.2:1b',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}], {'role': 'user', 'content': 'Why is the sky blue?'}
    stream=True,
)

for chunk in newstream:
  print(chunk['message']['content'], end='', flush=True)
```

```python
import ollama

messages = []

def send(chat, version):
  messages.append(
    {
      'role': 'user',
      'content': chat,
    }
  )
  stream = ollama.chat(model=version, 
    messages=messages,
    stream=True,
  )

  response = ""
  for chunk in stream:
    part = chunk['message']['content']
    print(part, end='', flush=True)
    response = response + part

  messages.append(
    {
      'role': 'assistant',
      'content': response,
    }
  )

def investigate(starter, predicate, model, times):
   messages.clear()
   send(starter, model)
   for i in range(times):
      send("Repeat that with more {}".format(predicate), model)

investigate("Imagine you are a waitress at a cafe in Western Europe. Tell me about your day.", "manic", "llama3.2:1b", 5)
```
